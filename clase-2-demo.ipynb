```python
import altair as alt
import pandas as pd 
from vega_datasets import data as vega_data
data = vega_data.gapminder()
```

```python
data.head(10)
```

```python
latam = data[data['cluster'] == 3]
```

```python
latam.head(50)
```

## Barras

```python
alt.Chart(latam).mark_circle().encode(
    y= alt.Y("country", title=""),
    x= alt.X("life_expect", title = "Esperanza de Vida"),
    color = alt.Color('fertility', scale = alt.Scale(scheme = "reds")),
)
```

```python
alt.Chart(latam).mark_circle().encode(
    y= alt.Y("country", title=""),
    x= alt.X("life_expect", title = "Esperanza de Vida"),
    color = alt.Color('fertility', title = "Fertilidad", scale = alt.Scale(scheme = "reds")),
    size = alt.Size("pop", title = "Población")
).properties(
 title = "Relación entre la esperanza de vida y la fertilidad",
)
```
